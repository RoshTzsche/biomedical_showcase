```python
import sys
import os
import torch
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

v1 = torch.tensor([[1., 0.], [1., 0.]])
v2 = torch.tensor([[0., 1.], [0., 1.]])
angle = torch.atan2(v1,v2)
```

```python
def convert(rad):
    deg = rad * 180/np.pi()
    return deg

degree = convert(angle)
```
